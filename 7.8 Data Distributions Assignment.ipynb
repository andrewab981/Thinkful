# Data Distributions Assignment
Use python to create some 6 random datasets in NumPy. Then use statistical knowledge to summarize them.

#### 1. Generate a random variable with 100 datapoint

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
n = 100
size = 100
p = .5
std = 30
mean = 75
scale = 2
```

#### Graph 6 Numpy Distributions

```python
binomial = np.random.binomial(n, p, size)
plt.hist(binomial)

plt.axvline(binomial.mean(), color = "orange", linestyle = 'solid', linewidth = 2)
plt.axvline(binomial.mean() + binomial.std(), color = "orange", linestyle = 'dashed', linewidth = 2)
plt.axvline(binomial.mean() - binomial.std(), color = "orange", linestyle = 'dashed', linewidth = 2)

plt.show()
```

```python
wald = np.random.wald(mean, scale, size)
plt.hist(wald)

plt.axvline(wald.mean(), color = "orange", linestyle = "solid", linewidth = 2)
plt.axvline(wald.mean() + wald.std(), color = "orange", linestyle = "dashed", linewidth = 2)
plt.axvline(wald.mean() - wald.std(), color = "orange", linestyle = "dashed", linewidth = 2)

plt.show()
```

```python
rayleigh = np.random.rayleigh(scale, size)

plt.hist(rayleigh)

plt.axvline(rayleigh.mean(), color = "orange", linestyle = "solid", linewidth = 2)
plt.axvline(rayleigh.mean() + rayleigh.std(), color = "orange", linestyle = "dashed", linewidth = 2)
plt.axvline(rayleigh.mean() - rayleigh.std(), color = "orange", linestyle = "dashed", linewidth = 2)

plt.show()
```

```python
logseries = np.random.logseries(p, size)

plt.hist(logseries)

plt.axvline(logseries.mean(), color = "orange", linestyle = "solid", linewidth = 2)
plt.axvline(logseries.mean() + logseries.std(), color = "orange", linestyle = "dashed", linewidth = 2)
plt.axvline(logseries.mean() - logseries.std(), color = "orange", linestyle = "dashed", linewidth = 2)

plt.show()
```

```python
cauchy = np.random.standard_cauchy(size)

plt.hist(cauchy)

plt.axvline(cauchy.mean(), color = "orange", linestyle = "solid", linewidth = 2)
plt.axvline(cauchy.mean() + cauchy.std(), color = "orange", linestyle = "dashed", linewidth = 2)
plt.axvline(cauchy.mean() - cauchy.std(), color = "orange", linestyle = "dashed", linewidth = 2)

plt.show()
```

```python
exponential = np.random.exponential(scale, size)

plt.hist(exponential)

plt.axvline(exponential.mean(), color = "orange", linestyle = "solid", linewidth = 2)
plt.axvline(exponential.mean() + exponential.std(), color = "orange", linestyle = "dashed", linewidth = 2)
plt.axvline(exponential.mean() - exponential.std(), color = "orange", linestyle = "dashed", linewidth = 2)

plt.show()
```

#### Create 2 normally distirbuted variables
Add the 2 variables together, and generate histogram. 

```python
mean1 = 5
std1 = .5
nom1 = np.random.normal(mean1, std1, size)

mean2 = 10
std2 = 1
nom2 = np.random.normal(mean2, std2, size)

nom3 = nom1 + nom2
plt.hist(nom3)

plt.axvline(nom3.mean(), color = "orange", linestyle = "solid", linewidth = 2)
plt.axvline(nom3.mean() + nom3.std(), color = "orange", linestyle = "dashed", linewidth = 2)
plt.axvline(nom3.mean() - nom3.std(), color = "orange", linestyle = "dashed", linewidth = 2)

plt.show()
```
